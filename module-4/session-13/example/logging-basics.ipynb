# Setup
1. Only run the following cell to save the code of the basic operations of the calculator.

````python
%%writefile calculadora.py
def obtener_fracciones(frac_str):
    if isinstance(frac_str, int) or isinstance(frac_str, float):
        return frac_str

    if "/" in frac_str:
        try:
            return float(frac_str)
        except ValueError:
            num, denom = frac_str.split("/")
            try:
                leading, num = num.split(" ")
                whole = float(leading)
            except ValueError:
                whole = 0
            frac = float(num) / float(denom)
            return whole - frac if whole < 0 else whole + frac
    else:
        return float(frac_str)


def sum(a, b):
    sumando_a = obtener_fracciones(a)
    sumando_b = obtener_fracciones(b)
    return sumando_a + sumando_b


def subtract(a, b):
    minuendo = obtener_fracciones(a)
    sustraendo = obtener_fracciones(b)
    return minuendo - sustraendo


def multiply(a, b):
    multiplicando = obtener_fracciones(a)
    multiplicador = obtener_fracciones(b)
    return multiplicando * multiplicador


def divide(a, b):
    dividendo = obtener_fracciones(a)
    divisor = obtener_fracciones(b)
    try:
        return dividendo / divisor
    except ZeroDivisionError:
        return "Division entre cero"
````

# Debug con logging


To debug correctly, the `logging` dependency is used. And is used as seen in the code: `logging.debug ("My message) `.


````python
%%writefile debug_logging.py
import logging
import calculadora as calc

def get_sum(v1,v2):
  return calc.sum(v1,v2)

def get_subtract(v1,v2):
  return calc.subtract(v1,v2)

def get_division(v1,v2):
  return calc.divide(v1,v2)

def get_multiplication(v1,v2):
  return calc.multiply(v1,v2)

num_1 = 20
num_2 = 10

add_result = get_sum(num_1, num_2)
logging.debug(f"Add: {num_1} + {num_2} = {add_result}")

sub_result = get_subtract(num_1, num_2)
logging.debug(f"Sub: {num_1} - {num_2} = {sub_result}")

mul_result = get_multiplication(num_1, num_2)
logging.debug(f"Mul: {num_1} * {num_2} = {mul_result}")

div_result = get_division(num_1, num_2)
logging.debug(f"Div: {num_1} / {num_2} = {div_result}")
````

Nothing happens, Logging's default configuration is WARNING. You will only throw WARNINGS or CRITICAL.

It is necessary to change the filter level.

````python
%%writefile debug_logging.py
import logging 
import calculadora as calc


logging.basicConfig(level=logging.DEBUG) 

def get_sum(v1,v2):
  return calc.sum(v1,v2)

def get_subtract(v1,v2):
  return calc.subtract(v1,v2)

def get_division(v1,v2):
  return calc.divide(v1,v2)

def get_multiplication(v1,v2):
  logging.debug(f"v1: {v1}")
  logging.debug(f"v2: {v2}")
  return calc.multiply(v1,v2)

num_1 = 20
num_2 = 10

add_result = get_sum(num_1, num_2)
logging.debug(f"Add: {num_1} + {num_2} = {add_result}")

sub_result = get_subtract(num_1, num_2)
logging.debug(f"Sub: {num_1} - {num_2} = {sub_result}")

mul_result = get_multiplication(num_1, num_2)
logging.info(f"Mul: {num_1} * {num_2} = {mul_result}")

div_result = get_division(num_1, num_2)
logging.debug(f"Div: {num_1} / {num_2} = {div_result}")
logging.warning("Mensaje de WARNING")
logging.error("Mensaje de ERROR")
````

# Splitting logs
A good practice is to have a log for each module.

To do this, first you have to create a `logger` instance, and configure it to take the name of the module as follows:

```python
my_logger = logging.getLogger(__name__)
```

````python
%%writefile employee.py
import logging

logger = logging.getLogger(__name__) # Indicamos que tome el nombre del modulo
logger.setLevel(logging.DEBUG) # Configuramos el nivel de logging

formatter = logging.Formatter('%(asctime)s:%(name)s:%(module)s:%(levelname)s:%(message)s') # Creamos el formato

file_handler = logging.FileHandler('employee.log') # Indicamos el nombre del archivo

file_handler.setFormatter(formatter) # Configuramos el formato

logger.addHandler(file_handler) # Agregamos el archivo

class Employee:
    """A sample Employee class"""

    def __init__(self, first, last):
        self.first = first
        self.last = last

        # Cambiar el logging por logger
        logger.info(f"Created Employee: {self.fullname} - {self.email}")

    @property
    def email(self):
        return f"{self.first}.{self.last}@email.com"

    @property
    def fullname(self):
        return f"{self.first} {self.last}"


emp_1 = Employee("John", "Smith")
emp_2 = Employee("Corey", "Schafer")
emp_3 = Employee("Jane", "Doe")
````
