```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://www.openml.org/data/get_csv/16826755/phpMYEkMl')
df.shape
```

```python
df.head()
```

```python
df.replace('?', np.nan, inplace=True)
```

```python
df.dtypes
```

Observe that age and fare are object types due to nan values

```python
df['age'] = df['age'].astype('float')
df['fare'] = df['fare'].astype('float')
```

```python
target = 'survived'
num_vars = [col for col in df.columns if df[col].dtype != object and col != target]
cat_vars = [col for col in df.columns if df[col].dtype == object]
```

```python
len(num_vars) + len(cat_vars) + 1 == df.shape[1]
```

```python
for col in df.columns:
    print(col)
    print('='*len(col))
    print(df[col].value_counts(dropna=False))
    print()
```

```python
df.isna().sum()/df.shape[0]
```

```python
def extract_title(input_string):
    return input_string.split(',')[1].split('.')[0].strip()

df['title'] = df['name'].apply(extract_title)
```
